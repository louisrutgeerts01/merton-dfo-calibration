```python
# realistic params (example)
mu = 0.05         # annual drift
sigma = 0.2       # annual vol
lambd = 1.0       # jumps/year
mu_j = -0.1       # mean log-jump
sigma_j = 0.2     # std log-jump
dt = 1/252
n_paths, n_steps = 1000, 252
S0 = 100.0

kappa = np.exp(mu_j + 0.5 * sigma_j**2) - 1.0  # IMPORTANT
S = np.zeros((n_paths, n_steps + 1))
print(len(S[0, :]))
S[:, 0] = S0

for i in range(n_steps):
    Z = np.random.standard_normal(n_paths)
    K = np.random.poisson(lambd * dt, size=n_paths)
    jump_log = K * mu_j + np.random.standard_normal(n_paths) * np.sqrt(K) * sigma_j
    drift_log = (mu - lambd * kappa - 0.5 * sigma**2) * dt  # drop "- lambd*kappa" for historical
    log_ret = drift_log + sigma * np.sqrt(dt) * Z + jump_log

    # Optional: guard against extreme underflow (debugging)
    # log_ret = np.maximum(log_ret, -50)

    S[:, i+1] = S[:, i] * np.exp(log_ret)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_theme(style="whitegrid")

# Assume we already have S and n_steps from the simulation
times = np.linspace(0, n_steps*dt, n_steps+1)

# Plot the first simulated path
sns.lineplot(x=times, y=S[0, :])
plt.xlabel("Time (years)")
plt.ylabel("Price")
plt.title("Merton Jump Diffusion - Example Path")
plt.show()
```
